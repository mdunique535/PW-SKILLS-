Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy():
    for exception_class in Exception.__subclasses__():
        print(exception_class.__name__)
        for subclass in exception_class.__subclasses__():
            print("  -", subclass.__name__)

print_exception_hierarchy()
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

```python
"""
The ArithmeticError class in Python is a base class for exceptions that occur during arithmetic operations. 
It serves as a parent class for several specific arithmetic-related exceptions. 
Here are two examples of errors defined in the ArithmeticError class:
"""

# ZerodivisionError

num3 = 10
num4 = 0

try:
    result3 = num3/num4
    print("result is :" , result3)
    
except ZeroDivisionError as z:
    print(z)
```

```python
# OverflowError

try:
    result = 2 ** 1000000
    print("Result:", result)
    
except ValueError as v:
    print(v)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
"""
The LookupError class in Python is a base class for exceptions that occur when a lookup or indexing operation fails. 
It serves as a parent class for specific lookup-related exceptions. 
Let's explain two examples of errors defined in the LookupError class: KeyError and IndexError.
"""

# KeyError

my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']
    print("Value:", value)
except KeyError:
    print("Error: Key 'd' does not exist in the dictionary.")
```

```python
# IndexError

my_list = [1,2,3,4,5]

try :
    
    i = my_list[10]
    print("INdex number is : " , i)
    
except IndexError as I:
    print(I)
```
